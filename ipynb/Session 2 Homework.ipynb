```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
X = [10, 20, 100, 1000]
X=np.array(X)
pmf = [0.4, 0.3, 0.2, 0.1]
pmf=np.array(pmf)

cdf = np.cumsum(pmf)
# Bảng phân phối xác suất
pmftable = np.column_stack((X, pmf))
print('Bảng trọng số: \n', pmftable)
# Vẽ đồ thị PMF, CDF
PMF = sns.barplot(X,pmf)
PMF.set(xlabel='gia tri cua X', ylabel='PX(x)')
plt.show()

CDF = sns.barplot(X, cdf)
CDF.set(xlabel='gia tri cua X', ylabel='FX(x)')
plt.show()
```

```python
# Tìm xác suất để biến ngẫu nhiên nhỏ hơn 50:
print('Xac suat de bien ngau nhien nho hon 50: ', pmf[X<=50].sum())
```

```python
# Tìm giá trị trung bình, phương sai, độ lệch tiêu chuẩn, trung vị, skewness:
mu = np.sum(X*pmf)
var = np.sum((X-mu)**2*pmf)
sd = np.sqrt(var)
median = X[np.min(np.where(cdf>=0.5))]
skew = np.sum(((X-mu)/sd)**3*pmf)
print('Ky vong: ', mu)
print('Phuong sai: ', var)
print('Do lech chuan: ', sd)
print('Trung vi: ', median)
print('Skewness: ', skew)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets
iris = datasets.load_iris()
#print(iris.keys())
x = iris.data
y = iris.target

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)
```

```python
# Tìm bảng trọng số của y, y_train, y_test
giatri, tanso_y = np.unique(y, return_counts=True)
_, tanso_ytrain = np.unique(y_train, return_counts=True)
_, tanso_ytest = np.unique(y_test, return_counts=True)
pmf_y = tanso_y/len(y)
pmf_ytrain = tanso_ytrain/len(y_train)
pmf_ytest = tanso_ytest/len(y_test)

print('Bảng trọng số của y: \n',np.column_stack((giatri, pmf_y)))
print('Bảng trọng số của y_train: \n',np.column_stack((giatri, pmf_ytrain)))
print('Bảng trọng số của y_test: \n',np.column_stack((giatri, pmf_ytest)))
```

```python
# Vẽ đồ thị hàm trọng số PMF của của y, y_train, y_test
PMF_y = sns.barplot(giatri,pmf_y)
PMF_y.set(xlabel='gia tri loai hoa, y', ylabel='P_Y(y)', title='Đồ thị hàm trọng số của y')
plt.show()
PMF_ytrain = sns.barplot(giatri,pmf_ytrain)
PMF_ytrain.set(xlabel='gia tri loai hoa, y', ylabel='P_Ytrain(y)', title='Đồ thị hàm trọng số của y_train')
plt.show()
PMF_ytest = sns.barplot(giatri,pmf_ytest)
PMF_ytest.set(xlabel='gia tri loai hoa, y', ylabel='P_Ytest(y)',title='Đồ thị hàm trọng số của y_test')
plt.show()
```

```python
# Vẽ đồ thị hàm phân phối xác suất CDF của y, y_train, y_test
cdf_y = np.cumsum(pmf_y)
cdf_ytrain = np.cumsum(pmf_ytrain)
cdf_ytest = np.cumsum(pmf_ytest)

CDF = sns.barplot(giatri, cdf_y)
CDF.set(xlabel='gia tri loai hoa, y', ylabel='F_Y(y)',title='Đồ thị hàm phân phối xác suất của y')
plt.show()

CDF = sns.barplot(giatri, cdf_ytrain)
CDF.set(xlabel='gia tri loai hoa, y', ylabel='F_Ytrain(y)',title='Đồ thị hàm phân phối xác suất của y_train')
plt.show()

CDF = sns.barplot(giatri, cdf_ytest)
CDF.set(xlabel='gia tri loai hoa, y', ylabel='F_Ytest(y)',title='Đồ thị hàm phân phối xác suất của y_test')
plt.show()
```
